```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['price'], bins=30 ,color='steelblue',edgecolor='black')
plt.title('Price Distribution',fontsize=16,fontweight='bold')
plt.xlabel("Price",fontweight='bold')
plt.ylabel("Count",fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(df['mileage'],df['price'],edgecolor='black')
plt.title('Mileage vs Price', fontsize=16, fontweight='bold')
plt.xlabel('Mileage', fontsize=12, fontweight='bold')
plt.ylabel('Price', fontsize=12, fontweight='bold')
plt.show()
```

```python
df.boxplot(column='price', by='fuelType', figsize=(10,6))
plt.title('Price by Fuel Type',fontsize=16,fontweight='bold')
plt.suptitle('')
plt.xlabel('Fuel Type',fontweight='bold')
plt.ylabel('Price',fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df.groupby('transmission')['price'].mean().plot(kind='bar')
plt.title('Average Price by Transmission',fontsize=16,fontweight='bold')
plt.xlabel('Transmission',fontweight='bold')
plt.ylabel('Average Price',fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['engineSize'], df['price'],edgecolor='black')
plt.title('Engine Size vs Price',fontsize=16,fontweight='bold')
plt.xlabel('Engine Size',fontweight='bold')
plt.ylabel('Price',fontweight='bold')
plt.show()


```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['year'], df['price'],edgecolor='black')
plt.title('Year vs Price',fontsize=16,fontweight='bold')
plt.xlabel('Year',fontweight='bold')
plt.ylabel('Price',fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['mpg'], df['price'],edgecolor='black')
plt.title('MPG vs Price',fontsize=16,fontweight='bold')
plt.xlabel('MPG',fontweight='bold')
plt.ylabel('Price',fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(df['price'])
plt.title('Price Outliers',fontsize=16,fontweight='bold')
plt.show()
```
